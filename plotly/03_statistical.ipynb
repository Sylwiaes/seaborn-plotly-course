### Histogram

```
df = load_dataset('mpg')
df.head()
```

```
df.describe()
```

```
df.origin.value_counts() #badamy rozkład zmiennej origin
```

```
px.histogram?
```

```
import plotly.express as px

px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg') #nbins-liczba słupków
```

```
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density') #histnorm-parametr, który zwraca gęstość prawdopodobieństwa
#na osi y mamy gęstość prawdopodobieństwa
```

```
px.histogram(df, x='mpg', nbins=50, title='Rozkład zmiennej mpg', histnorm='probability density',
            facet_row='origin', color='origin')
#facet_row-dzieli wykres w zależności od podanego parametru, dzieli wierszowo
#facet_col-dzieli kolumnowo
```

```
dff = df[(df.origin == 'usa') | (df.origin == 'europe')] #warunkowe wycinanie
px.histogram(dff, x='mpg', color='origin', marginal='violin', title='Rozkład zmiennej mpg (USA vs. Europe)', opacity=0.7)
#opacity-przezroczystość
#wniosek z poniższego wykresu-na europejskich samochodach można wykonać więcej km, niż na samochodach amerykańskich
```

```
px.histogram(df, x='origin')
```

```
import plotly.figure_factory as ff
#tworzymy wykres dystrybucji dla zmiennej mpg
#funkcja gęstości to linia na wykresie
fig = ff.create_distplot([df.mpg], group_labels=['mpg'])
fig.show()
```

```
fig = ff.create_distplot([df.mpg, df.acceleration], group_labels=['mpg', 'Acceleration'])
fig.show()
#acceleration-przyspieszenie
```

```
import numpy as np

x1 = np.random.randn(1000)
x2 = np.random.randn(1000) + 2 #wygenerowanie rozkładu normalnego i przesunięcie wykresu o 2

ff.create_distplot([x1, x2], ['Rozkład Normalny ~ N(0, 1)', 'Rozkład Normalny ~ N(2, 1)'], curve_type='normal',
                  bin_size=0.2, colors=['slategray', 'magenta'])
#rozkład normalny 0,1 i 2,1 i o odchyleniu 1. Typ krzywej-curve_type-normalna
#bin_size-rozmiar słupka, parametr colors-lista kolorów
```

### Histogram 2D

```
df = load_dataset('mpg')
df.head()
```

```
import plotly.graph_objects as go

fig = go.Figure(go.Histogram2d(
    x=df.mpg,
    y=df.acceleration,
    nbinsx=30,
    nbinsy=30
))
#nbinsx-kontroluje podziały względem osi x
#nbinsy-kontroluje podziały wzgledem osi y

#dodajemy wykres punktowy do histogramu:
fig.add_trace(go.Scatter(
    x=df.mpg,
    y=df.acceleration,
    mode='markers',
    marker={
        'symbol': 'x',
        'opacity': 0.7,
        'color': 'white',
        'size': 8,
        'line': {
            'width': 1
        }
    }
))

fig.show()
```
