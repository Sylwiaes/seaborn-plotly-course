```
import matplotlib.pyplot as plt
import seaborn as sns #biblioteka seaborn jest zbudowana na biliotece matplotlib; nakładka na biblioteke matplotlib, która pozwala tworzyć wykresy statystyczne
import pandas as pd
import numpy as np
```

```
#tworzymy przykładowe dane, z którymi będziemy pracować
x = np.linspace(0, 20, 100) #linspace generuje dane od 0-20 równo rozłożonych 100 wartości
y = np.random.randn(100, 4) #dane z rozkładu normalnego: 100 próbek i 4 zmienne
y = np.cumsum(y, axis=0) #suma skumulowana po kolumnach, dodanie poprzednie kolumny dla osi 0
```

```
df = sns.load_dataset('tips') #ładujemy dane, z którymi będziemy pracować
df.head() #wyświetlamy 5 pierwszych wierszy
```

```
df.info() #sprawdzamy typy danych
```

```
df.describe() #wyświetlamy statystyki danych
```

```
df.describe(include=['category']) #wyświetlamy statystyki zmiennych kategorycznych, które zostały określone wyżej w Dtype
```

```
sns.set_context('paper')
sns.relplot(data=df, x='total_bill', y='tip') #tworzymy wykres relplot wraz z ramką danych, oś x i y
#Wykresy relplot - wykresy relacyjne:
```

Wykresy relplot - podstawowe parametry:

*   data – wskazujemy zbiór danych
*   x, y – wskazujemy dane, które mają być umieszczone na osi x oraz y
*   hue – jeżeli chcemy, aby dane były kolorystycznie różne, w zależności od wartości zmiennej, to tutaj ją podajemy
*   Ecol, row – w ilu kolumnach i wierszach mają się wyświetlić wykresy
*   kind – typ wykresu – czy liniowy, czy z punktami
*   aspect – szerokość wykresu


```
sns.set(style='darkgrid', palette='deep', font_scale=1) #zmieniamy styl naszego wykresu, ustawiamy parametry
sns.relplot(data=df, x='total_bill', y='tip') 
```

```
sns.set(palette='pastel', font_scale=1.2)
sns.relplot(data=df, x='total_bill', y='tip')
```

```
sns.set(font_scale=1.2)
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', sizes=(10, 200), palette='viridis',
           style='size') #parametr hue-pozwala skategoryzować dane na wykresie względem podanej kolumny
           #parametr size kontroluje rozmiar pojedynczego punktu
           #parametr sizes zmniejsza/zwiększa rozmiar kropek; (10, 200) oznacza, że będziemy mieć rozmiary najmniejszy i największy
           #parametr palette oznacza paletę kolorów
           #parametr style pozwala kontrolować style naszych kropek
```

```
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', sizes=(10, 200), palette='viridis',
           col='time') #dzielimy powyższy wykres na 2 wykresy, w zależności od pory czasu - col='time'
```

```
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', sizes=(10, 200), palette='viridis',
           row='time') #przedstawienie wykresu w pionie prz pomocy parametry row
```

```
sns.relplot(data=df, x='total_bill', y='tip', size='size', hue='size', sizes=(10, 200), palette='viridis',
           col='time', row='smoker') #tworzymy siatki wykresów - parametr col i row
```

### Rozkłady - wykresy

```
sns.set(style='darkgrid') #styl darkgrid pozwala na wydrukowanie siatki na wykresie

df = np.random.multivariate_normal(mean=[0, 1], cov=[(1, -0.4), (-0.4, 1)], size=200) #generujemy rozkład normalny wielowymiarowy o średniej 0,1 i macierzy kowariancji (czyli macierz
#symetryczna)
df = pd.DataFrame(df, columns=['x', 'y'])
df.head(3)
```

```
df.describe() #sprawdzamy statystyki
```

```
(sns.jointplot(data=df, x='x', y='y')) #joinplot-wykres rozrzutu rozkładu normalnego
#wykres przedstawia histogramy z obu stron, dla niewielu zmiennych jest on bardziej płaski, dla większej ilości zmiennych jest wyższy
```

```
sns.set(style='white')
sns.jointplot(data=df, x='x', y='y', kind='hex') #hex-wykres plastry miodu, warto tu zastosować styl white
```

```
sns.jointplot(data=df, x='x', y='y', kind='kde') #kde-wykres estymacji naszych danych
```

```
g = sns.jointplot(data=df, x='x', y='y', kind='kde', n_levels=15, cmap='viridis') #parametr n_levels dodaje liczbę poziomów
g.plot_joint(plt.scatter, c='w', s=30, marker='+') #dodajemy wykres punktowy
```

### Pairplot - wykresy relacji parami

```
df = sns.load_dataset('iris')
df.head(3)
```

```
sns.set()
sns.pairplot(data=df) #pairplot-pozwala na szybkie sprawdzenie zmiennych numerycznych
#funkcja pairplot wybiera wszystkie dane, które są danymi numerycznymi i przedstawia je za pomoca wykresu
#zmienna species nie zostaje wzięta pod uwagę do wykresu
```

```
sns.pairplot(data=df, hue='species') #dodajemy parametr hue i sprawdzamy jak puntky rozkładają się w zależności od gatunku
#klasy czerwona i zielona są bardzo do siebie podobne, aby rozdzielić te 2 zbiory, patrzymy na wykresy i staramy się wbrać takie, w których te dane można łatwo rozdzielić
#budujemy model, który przewiduje klasę na podstawie 4 danych numerycznych
```

```
sns.pairplot(data=df, hue='species', palette='pastel')
```

```
g = sns.PairGrid(df, hue='species') #tworzymy pusty pairplot
g.map_diag(sns.kdeplot) #mapujemy nazwę wykresu, który chcemy użyć na naszej macierzy diagonalnej (na przekątnej); przekazujemy funkcję kde(Kernel destination plott)
g.map_offdiag(sns.kdeplot, n_levels=6) #mapujemy funkcję poza przekatną
```
